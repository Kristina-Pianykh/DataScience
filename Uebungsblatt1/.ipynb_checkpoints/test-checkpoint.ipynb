```python
def get_mean(lst: List[float]) -> float:
    accumulator = 0.0
    for item in lst:
        accumulator += item
    return accumulator / float(len(lst))

def get_mean(lst: List[float]) -> float:
    accumulator = 0.0
    for item in lst:
        accumulator += item
    return accumulator / float(len(lst))

def get_stddev(lst: List[float]) -> float:
    accumulator = 0.0
    mean = get_mean(lst)

    for item in lst:
        accumulator += (item - mean)**2

    variance = accumulator / float(len(lst) - 1)
    return sqrt(variance)


def get_confidence_interval(bootstrap_medians: List[float], p: float) -> Tuple[float, float]:
    sorted_medians = sorted(bootstrap_medians)
    within_interval = float(len(bootstrap_medians)) * p
    
    big_decimal_part = (within_interval - floor(within_interval)) >= 0.5
    if big_decimal_part:
        within_interval = ceil(within_interval)
    else:
        within_interval = floor(within_interval)
    
    to_remove = floor((len(bootstrap_medians) - within_interval) / 2)
    lower_idx = 0 + to_remove
    upper_idx = len(sorted_medians) - to_remove
    return (sorted_medians[lower_idx], sorted_medians[upper_idx-1])
```

```python
from pathlib import Path

# Die Eingabedatei muss im gleichen Verzeichnis liegen wie die Skriptdatei exercise_2
input_file = Path("todesursachen.csv")
data: pd.DataFrame = pd.read_csv(input_file, sep=";", header=0)
```

```python
def aufgabe_b2() -> Tuple[float, float]:
    """
    Ertrinken im Durchschnitt mehr Kinder bzw. Heranwachsende (< 15 Jahre) oder mehr Menschen im
    Rentenalter (>= 65 Jahre) pro Jahr?

    Als Rückgabe wird ein Pair von float-Zahlen erwartet, welche die beiden Durchschnittswerte der jeweiligen
    Personengruppen angeben:

    (<Durchschnitt-Kinder (float)>, <Durchschnitt-Renter*innen (float)>)

    :return: (<Durchschnitt-Kinder (float)>, <Durchschnitt-Renter*innen (float)>)
    """
    kids_age_groups = [
        "unter 1 Jahr",
        "1 bis unter 15 Jahre"
    ]
    old_age_groups = [
        "65 bis unter 70 Jahre",
        "70 bis unter 75 Jahre",
        "75 bis unter 80 Jahre",
        "80 bis unter 85 Jahre",
        "85 Jahre und mehr"
    ]
    drowned_set = data[(data["Todesursache"] == "Ertrinken und Untergehen")]
    drowned_kids = drowned_set[drowned_set["Altersgruppe"].isin(kids_age_groups)]["Anzahl"].mean()
    drowned_old_fucks = drowned_set[drowned_set["Altersgruppe"].isin(old_age_groups)]["Anzahl"].mean()
    return (float(drowned_kids), float(drowned_old_fucks))
```

```python
aufgabe_b2()
```
